```python
name=input("enter std. name:")
a=int(input("enter 1st sub. Mark:"))
b=int(input("enter 2nd sub. Mark:"))
c=int(input("enter 3rd sub. Mark:"))
d=int(input("enter 4th sub. Mark:"))
e=int(input("enter 5th sub.Mark:"))
mark=[a,b,c,d,e]
def add():
     total=a+b+c+d+e

     avg=total/5

     if avg>=75 and avg<=100:
         grade= "A"
     elif avg>=50 and avg<75:
         grade= "B"
     elif avg<50 and avg>34:
         grade= "C"
     else:
         grade= "A"

     return total, avg , grade

total,avg, grade=add()

student = {"Name": name ,
           "Mark": mark,
           "Total": total,
           "Average": avg,
           "Grade": grade}
print ("-------student report---------")
print (f"Name    : {student["Name"]}")
print (f"Mark    : {student["Mark"]}")
print (f"Total   : {student["Total"]}")
print (f"Average : {student["Average"]}")
print (f"Grade   : {student["Grade"]}")
```
